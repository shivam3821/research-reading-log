```python
s = "12345678"
start = len(s)-1
end = -1
step = -1
index = -1

for i in range(start,end,step):
    if int(s[i])%2==1:
        index = i
        break

ss = ""

if(index!=-1):
    for i in range(0,index+1):
        ss = ss + s[i]
print(ss)
```

```python
def anagram(a,b):
    if len(a)!=len(b):
        return False
    
    char_freq = {}

    for i in a:
        char_freq[i] = char_freq.get(i,0)+1
    
    for i in b:
        if i not in char_freq:
            return False
        else:
            if char_freq[i]==0:
                return False
            else:
                char_freq[i]-=1
    
    return True
```

```python
a = 'anagram'
b = 'ngaram'

anagram(a,b)
```
