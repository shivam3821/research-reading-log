## Sparse Retrieval vs Dense Retrieval

| Sparse Retrieval | Dense Retrieval |
|-----------------|------------------|
| based on Sparse vectors (most entries are zeroes) few positions contatins words weights | Dense vectors (embeddings) created by neural network | 
| exact word matching | captures semantic meaning | 
| convert documents to term vector, assign weights (TF-IDF etc) | convert documents to embedding using a model, store into vector database | 
| Retrieve using exact word match | Retrieve using similarity algos (cosine similarity) |
